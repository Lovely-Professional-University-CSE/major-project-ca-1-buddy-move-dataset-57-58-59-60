```python
import numpy as np
import matplotlib.pyplot as plt
x= np.arange(-10,10,1,dtype='int32')
print(x)
def identity(x):
    return(x)
def plot_graph(x,y,name):
    plt.plot(x,y)
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.title(name)
y=identity(x)
plot_graph(x,y,'identity')
```

```python
def sigmoidal(x):
    return(1/1+np.exp(-x))
y=sigmoidal(x)
plot_graph(x,y,'sigmoidal')
```

```python
def tanh(x):
    return(2/1+np.exp(-2*x))-1
y=tanh(x)
plot_graph(x,y,'tanh')
```

```python
def archtanh(x):
    return np.arctan(x)

y = archtanh(x)
plot_graph(x,y,'archtanh')
```

```python
def relu(x):
    if x>0:
        return x
    else:
        return 0

y = np.empty(len(x))
for i in range(len(x)):
    y[i] = relu(x[i])
plot_graph(x,y,'relu')
```
